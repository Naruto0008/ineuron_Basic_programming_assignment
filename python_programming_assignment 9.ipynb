```python
6. Write a Python program to print all pronic numbers between 1 and 100?
```

```python
def PronicNo(num):
    Pronic = False
    for i in range(1,num+1):
        if i*(i+1) == num:
            Pronic = True
            break
    return Pronic

print("Pronic numbers in range 1 to 100")
pronicno = []
for i in range(1,101):
    if PronicNo(i):
        pronicno.append(i)
print(pronicno)
```
